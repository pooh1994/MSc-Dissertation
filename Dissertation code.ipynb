```python

def load_csv_safely(path):
    encodings = ["utf-8-sig", "utf-8", "cp1252", "latin1", "utf-16"]
    seps = [None, ",", ";", "\t", "|"]  # None = infer (python engine)
    
    last_err = None
    for enc in encodings:
        for sep in seps:
            try:
                df_try = pd.read_csv(
                    path,
                    encoding=enc,
                    sep=sep,
                    engine="python",
                    on_bad_lines="skip"
                )
                # success criteria: has at least 2 columns OR at least 1 row
                if df_try.shape[1] > 1 or df_try.shape[0] > 0:
                    print("Loaded with encoding =", enc, "| sep =", repr(sep), "| shape =", df_try.shape)
                    return df_try
            except Exception as e:
                last_err = e
                continue
    raise last_err

df = load_csv_safely(path)
df.head()
```

```python
# Set a random seed to ensure reproducibility of capacity splits
np.random.seed(42)

# Calculate the total incoming textile weight across all batches
total_input = df["weight_kg"].sum()

# Define sorting centres
# Total sorting capacity is distributed unevenly to reflect real-world imbalances
sorting_centres = ["S1", "S2", "S3"]
sorting_capacity = {
    "S1": 0.40 * total_input,
    "S2": 0.35 * total_input,
    "S3": 0.25 * total_input
}

# Identify unique recycling methods present in the dataset
methods = sorted(df["recycling_method"].unique().tolist())

# Initialise dictionaries to store recycling facilities and their capacities
recyclers = {}
recycler_capacity = {}

# Create recycling facilities and assign capacities by method
for m in methods:
    # Create three facilities per recycling method
    # Example: mechanical -> M1, M2, M3
    facs = [f"{m[:1].upper()}{i}" for i in range(1, 4)]
    recyclers[m] = facs
    
    # Calculate the share of total input handled by this recycling method
    # This reflects historical load observed in the dataset
    method_share = (
        df.loc[df["recycling_method"] == m, "weight_kg"].sum() / total_input
    )
    
    # Define total capacity for this method based on its observed share
    method_total_cap = method_share * total_input
    
    # Split capacity unevenly across the three facilities
    # A Dirichlet distribution is used to introduce realistic imbalance
    splits = np.random.dirichlet([2, 2, 2], 1).flatten()
    for f, s in zip(facs, splits):
        recycler_capacity[f] = s * method_total_cap

# Display sorting centre capacities and recycling facility capacities
sorting_capacity, recycler_capacity

```

```python
# Assign each batch to a sorting centre
# The assignment probabilities are based on each centre’s share of total sorting capacity
sc_probs = (
    np.array([sorting_capacity[s] for s in sorting_centres]) /
    sum(sorting_capacity.values())
)
df["sorting_centre"] = np.random.choice(
    sorting_centres, size=len(df), p=sc_probs
)

# Assign each batch to a recycler facility based on its recycling method
# Each batch is routed only to facilities that match its method (e.g., mechanical -> M1/M2/M3)
df["recycler_facility"] = df["recycling_method"].apply(
    lambda m: np.random.choice(recyclers[m])
)

# Preview key routing fields to confirm the assignments look correct
df[["batch_id", "weight_kg", "sorting_centre", "recycling_method", "recycler_facility"]].head()

```

```python
# Import libraries needed for numerical operations and network flow modelling
import numpy as np
import networkx as nx

# Create a directed graph for the supply chain network
G = nx.DiGraph()

# Define source and sink nodes
# COLLECTION represents total incoming textile supply
# END represents the final destination (all material must reach here)
source = "COLLECTION"
sink = "END"

# Add source and sink nodes with temporary demand values (updated later)
G.add_node(source, demand=0)
G.add_node(sink, demand=0)

# Add sorting centre nodes
for s in sorting_centres:
    G.add_node(s, demand=0)

# Add recycler facility nodes (grouped by recycling method)
for m in methods:
    for r in recyclers[m]:
        G.add_node(r, demand=0)

# Define scaling to convert kg into integer flow units
# Here, 1 unit represents 100 kg to keep the flow problem tractable
scale = 100

# Add a small buffer to capacities to avoid infeasibility caused by rounding
buffer = 1.05  # 5% slack

# Convert total input into demand units (integer)
# I use floor to avoid overstating demand after scaling
demand_units = int(np.floor(total_input / scale))

# Assign supply (negative demand) to the source and demand (positive) to the sink
G.nodes[source]["demand"] = -demand_units
G.nodes[sink]["demand"] = demand_units

# Add edges from collection to sorting centres
# Capacities are scaled and buffered; transport cost is proportional to distance and weight
for s in sorting_centres:
    cap_units = int(np.ceil((sorting_capacity[s] * buffer) / scale))
    cost_per_unit = dist_cs[s] * transport_cost_rate * scale
    G.add_edge(source, s, capacity=cap_units, weight=float(cost_per_unit))

# Add edges from sorting centres to recycler facilities
# Total cost includes both transport cost and processing cost (based on recycling method)
for s in sorting_centres:
    for m in methods:
        for r in recyclers[m]:
            cap_units = int(np.ceil((recycler_capacity[r] * buffer) / scale))
            transport_cost = dist_sr[(s, r)] * transport_cost_rate * scale
            processing_cost = proc_cost_per_kg_method[m] * scale
            G.add_edge(
                s, r,
                capacity=cap_units,
                weight=float(transport_cost + processing_cost)
            )

# Add edges from recycler facilities to the end node
# I use a very large capacity so this final step does not constrain the solution
big_cap = int(1e12)
for m in methods:
    for r in recyclers[m]:
        G.add_edge(r, sink, capacity=big_cap, weight=0.0)

# Solve the min-cost flow problem on the network
# This finds the lowest-cost routing plan that satisfies all supply, demand, and capacity constraints
flow = nx.min_cost_flow(G)

# Calculate the total cost of the optimal flow solution
total_cost = nx.cost_of_flow(G, flow)

# Output the total cost of the optimised routing plan
total_cost
```

```python
# Import required components for preprocessing and modelling
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

# Identify categorical and numeric features in the dataset
categorical_features = X.select_dtypes(include="object").columns.tolist()
numeric_features = X.select_dtypes(exclude="object").columns.tolist()

# Create a preprocessing pipeline
# Categorical variables are one-hot encoded, while numeric variables are passed through
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),
        ("num", "passthrough", numeric_features),
    ]
)

# Define the Random Forest regression model
# The parameters are chosen to balance model complexity and generalisation
rf = RandomForestRegressor(
    n_estimators=400,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)

# Combine preprocessing and model into a single pipeline
# This ensures consistent transformations during training and testing
model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("rf", rf),
    ]
)
```

```python
# Define a cleaning function for the WRAP infrastructure dataset
# This prepares the data for analysis and ML use
def clean_wrap_infrastructure(path: Path) -> pd.DataFrame:
    
    # Load the dataset using latin1 encoding to avoid character issues
    df = pd.read_csv(path, encoding="latin1")
    
    # Standardise column names for consistency
    # I strip spaces, convert to lowercase, and replace spaces with underscores
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    
    # Drop any unnamed columns created during CSV import
    df = df.drop(columns=[c for c in df.columns if "unnamed" in c], errors="ignore")

    # Convert capacity-related columns to numeric where available
    # Invalid or non-numeric values are coerced to NaN
    for col in ["current_capacity", "2025_capacity", "2030_capacity"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    # Standardise all text columns
    # I convert values to strings, remove extra spaces, and apply lowercase formatting
    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].astype(str).str.strip().str.lower()

    # Drop rows without the target variable for modelling
    # Also remove duplicate rows to avoid double counting
    df = df.dropna(subset=["current_capacity"])
    df = df.drop_duplicates()

    # Return the cleaned infrastructure dataset
    return df
```

```python
# Min-cost flow network (optimised configuration)

def build_network(demand, capacities, costs):
    G = nx.DiGraph()

    # Source and sink
    G.add_node("COLLECTION", demand=-demand)
    G.add_node("OUTPUT", demand=demand)

    # Processing stages
    for node in ["SORTING", "MECH", "CHEM", "REUSE"]:
        G.add_node(node, demand=0)

    # Edges with capacity and cost
    for (u, v), cap in capacities.items():
        G.add_edge(u, v, capacity=cap, weight=costs[(u, v)])

    return G

# Example inputs
demand = 1000

capacities = {
    ("COLLECTION", "SORTING"): 1000,
    ("SORTING", "MECH"): 500,
    ("SORTING", "CHEM"): 200,
    ("SORTING", "REUSE"): 300,
    ("MECH", "OUTPUT"): 500,
    ("CHEM", "OUTPUT"): 200,
    ("REUSE", "OUTPUT"): 300,
}

costs = {
    ("COLLECTION", "SORTING"): 2,
    ("SORTING", "MECH"): 4,
    ("SORTING", "CHEM"): 6,
    ("SORTING", "REUSE"): 3,
    ("MECH", "OUTPUT"): 2,
    ("CHEM", "OUTPUT"): 2,
    ("REUSE", "OUTPUT"): 1,
}

# Solve optimisation
G = build_network(demand, capacities, costs)
opt_flow = nx.min_cost_flow(G)
opt_cost = nx.cost_of_flow(G, opt_flow)

opt_cost
```

```python
# Baseline routing using fixed priority rules

def baseline_flow(demand, capacities):
    flows = {}
    flows[("COLLECTION", "SORTING")] = demand

    remaining = demand
    for route in [("SORTING", "MECH"), ("SORTING", "REUSE"), ("SORTING", "CHEM")]:
        flow = min(capacities[route], remaining)
        flows[route] = flow
        remaining -= flow

    flows[("MECH", "OUTPUT")] = flows[("SORTING", "MECH")]
    flows[("REUSE", "OUTPUT")] = flows[("SORTING", "REUSE")]
    flows[("CHEM", "OUTPUT")] = flows[("SORTING", "CHEM")]

    return flows

def calculate_cost(flows, costs):
    return sum(flows[e] * costs[e] for e in flows)

baseline = baseline_flow(demand, capacities)
baseline_cost = calculate_cost(baseline, costs)

baseline_cost, opt_cost
```

```python
# Identify capacity bottlenecks

def identify_bottlenecks(G, flow):
    rows = []
    for u, v, data in G.edges(data=True):
        cap = data["capacity"]
        f = flow.get(u, {}).get(v, 0)
        rows.append({
            "Route": f"{u} → {v}",
            "Flow": f,
            "Capacity": cap,
            "Utilisation": f / cap
        })
    return pd.DataFrame(rows).sort_values("Utilisation", ascending=False)

bottlenecks = identify_bottlenecks(G, opt_flow)
bottlenecks.head()
```

```python
def generate_dissertation_outputs(df, top_n=10):
    out = {}  # Dictionary to store outputs

    # -----------------------------
    # A) Dataset evidence
    # -----------------------------
    out["dataset_rows"] = int(df.shape[0])  # Number of rows
    out["dataset_cols"] = int(df.shape[1])  # Number of columns
    out["columns"] = df.columns.tolist()    # Column names

    # Check if key columns exist
    key_cols = [
        "country", "region", "core_function", "main_activities",
        "current_capacity", "2025_capacity", "2030_capacity",
        "feedstock", "output_type", "output_composition", "date_last_updated"
    ]
    out["key_columns_present"] = {c: (c in df.columns) for c in key_cols}

    # Ensure capacity columns are numeric
    df2 = df.copy()
    for c in ["current_capacity", "2025_capacity", "2030_capacity"]:
        if c in df2.columns:
            df2[c] = pd.to_numeric(df2[c], errors="coerce")

    # -----------------------------
    # B) Descriptive capacity outputs
    # -----------------------------
    if "current_capacity" in df2.columns:
        out["current_capacity_total"] = float(df2["current_capacity"].sum(skipna=True))
        out["current_capacity_median"] = float(df2["current_capacity"].median(skipna=True))
        out["current_capacity_mean"] = float(df2["current_capacity"].mean(skipna=True))

    # Top feedstocks
    if "feedstock" in df2.columns and "current_capacity" in df2.columns:
        top_feedstock = (
            df2.dropna(subset=["feedstock", "current_capacity"])
              .groupby("feedstock")["current_capacity"].sum()
              .sort_values(ascending=False)
              .head(top_n)
        )
        out["top_feedstock_by_capacity"] = top_feedstock

    # Top core functions
    if "core_function" in df2.columns and "current_capacity" in df2.columns:
        top_functions = (
            df2.dropna(subset=["core_function", "current_capacity"])
              .groupby("core_function")["current_capacity"].sum()
              .sort_values(ascending=False)
              .head(top_n)
        )
        out["top_core_functions_by_capacity"] = top_functions

    # Top regions
    if "region" in df2.columns and "current_capacity" in df2.columns:
        top_regions = (
            df2.dropna(subset=["region", "current_capacity"])
              .groupby("region")["current_capacity"].sum()
              .sort_values(ascending=False)
              .head(top_n)
        )
        out["top_regions_by_capacity"] = top_regions

    # -----------------------------
    # C) OR / Prescriptive model evidence
    # -----------------------------
    out["predictive_model_present"] = ("best_model" in globals()) or ("model" in globals())
    out["optimisation_solution_present"] = ("sol" in globals())
    out["scenario_results_present"] = ("scenario_results" in globals())

    if "scenario_results" in globals():
        try:
            sr = scenario_results.copy()
            out["scenario_count"] = int(sr.shape[0])
            out["best_scenario"] = (
                sr.sort_values("objective_net_value", ascending=False).iloc[0].to_dict()
                if "objective_net_value" in sr.columns else None
            )
        except Exception:
            out["scenario_count"] = None
            out["best_scenario"] = None

    # -----------------------------
    # Print summary (report-ready)
    # -----------------------------
    print("DISSERTATION OUTPUT SUMMARY (WRAP + OR / PRESCRIPTIVE ANALYTICS)")
    print(f"Dataset size: {out['dataset_rows']} rows × {out['dataset_cols']} columns")

    print("\nKey columns present:")
    for k, v in out["key_columns_present"].items():
        print(f" - {k}: {v}")

    if "current_capacity_total" in out:
        print("\nCapacity overview (current_capacity):")
        print(f" - Total current capacity: {out['current_capacity_total']:.2f}")
        print(f" - Mean capacity: {out['current_capacity_mean']:.2f}")
        print(f" - Median capacity: {out['current_capacity_median']:.2f}")

    if "top_feedstock_by_capacity" in out:
        print(f"\nTop {top_n} feedstock streams by total current capacity:")
        print(out["top_feedstock_by_capacity"].to_string())

    if "top_core_functions_by_capacity" in out:
        print(f"\nTop {top_n} core functions by total current capacity:")
        print(out["top_core_functions_by_capacity"].to_string())

    if "top_regions_by_capacity" in out:
        print(f"\nTop {top_n} regions by total current capacity:")
        print(out["top_regions_by_capacity"].to_string())

    print("\nModeling artefacts detected in notebook:")
    print(f" - Predictive model present: {out['predictive_model_present']}")
    print(f" - Optimisation solution present: {out['optimisation_solution_present']}")
    print(f" - Scenario results table present: {out['scenario_results_present']}")

    if out.get("scenario_results_present"):
        print("\nScenario summary:")
        if out.get("scenario_count") is not None:
            print(f" - Number of scenarios evaluated: {out['scenario_count']}")
        if out.get("best_scenario") is not None:
            print(" - Best-performing scenario (by objective_net_value):")
            print(out["best_scenario"])

    print("\n(End of output summary)")

    return out

# Run the output generator
summary_outputs = generate_dissertation_outputs(df, top_n=10)
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import linprog

# ============================================================
# 0) Prepare WRAP infrastructure data
# ============================================================
df_nf = df.copy()
df_nf["current_capacity"] = pd.to_numeric(df_nf["current_capacity"], errors="coerce")
df_nf = df_nf.dropna(subset=["organisation", "region", "core_function", "current_capacity"]).copy()
df_nf = df_nf[df_nf["current_capacity"] > 0].copy()

# OPTIONAL: limit size for speed (increase if you want more realism)
df_nf = df_nf.sort_values("current_capacity", ascending=False).head(120).copy()

def map_stage(core_fn: str) -> str:
    s = str(core_fn).lower()
    if "sort" in s:
        return "sorting"
    if "reuse" in s:
        return "reuse"
    if "mechan" in s:
        return "mechanical"
    if "chem" in s:
        return "chemical"
    if "down" in s:
        return "downcycling"
    return "other"

df_nf["stage"] = df_nf["core_function"].apply(map_stage)

sorting_nodes = df_nf[df_nf["stage"] == "sorting"].copy()
proc_nodes = df_nf[df_nf["stage"].isin(["reuse","mechanical","chemical","downcycling","other"])].copy()

# If sorting nodes are too few, broaden selection
if len(sorting_nodes) < 3:
    sorting_nodes = df_nf[df_nf["core_function"].astype(str).str.lower().str.contains("sort", na=False)].copy()
    sorting_nodes["stage"] = "sorting"

# Assign node IDs
sorting_nodes = sorting_nodes.reset_index(drop=True)
proc_nodes = proc_nodes.reset_index(drop=True)
sorting_nodes["node_id"] = ["S"+str(i) for i in sorting_nodes.index]
proc_nodes["node_id"] = ["P"+str(i) for i in proc_nodes.index]

# ============================================================
# 1) Define total inflow supply (collection)
# ============================================================
total_sort_cap = sorting_nodes["current_capacity"].sum()
# Keep inflow <= sorting capacity to avoid forcing overflow at sorting stage
TOTAL_SUPPLY = 0.85 * total_sort_cap

# Supply distributed by region proportional to sorting capacity
sort_cap_by_region = sorting_nodes.groupby("region")["current_capacity"].sum()
regions = sort_cap_by_region.index.tolist()
region_supply = (sort_cap_by_region / sort_cap_by_region.sum() * TOTAL_SUPPLY).to_dict()

# Create one collection source node per region
source_nodes = [{"node_id": f"C_{r}", "region": r} for r in regions]

# ============================================================
# 2) Build arcs and costs (feasible via DISPOSAL sink)
# ============================================================
stage_cost = {"reuse": 1.0, "mechanical": 2.0, "chemical": 3.0, "downcycling": 1.5, "other": 2.5}
cross_region_penalty = 0.8

DISPOSAL_NODE = "DISPOSAL"
DISPOSAL_COST = 20.0  # large penalty so optimisation avoids it unless necessary
BIG_CAP = 1e12         # effectively unlimited

arcs = []

# (A) Collection -> Sorting (same region only)
for c in source_nodes:
    r = c["region"]
    sort_in_region = sorting_nodes[sorting_nodes["region"] == r]
    for _, srow in sort_in_region.iterrows():
        arcs.append({
            "from": c["node_id"],
            "to": srow["node_id"],
            "cost": 0.5,
            "cap": float(srow["current_capacity"])  # sorting capacity limits intake
        })

# (B) Sorting -> Processing
for _, srow in sorting_nodes.iterrows():
    for _, prow in proc_nodes.iterrows():
        base = stage_cost.get(prow["stage"], 2.5)
        penalty = cross_region_penalty if srow["region"] != prow["region"] else 0.0
        arcs.append({
            "from": srow["node_id"],
            "to": prow["node_id"],
            "cost": base + penalty,
            "cap": float(prow["current_capacity"])
        })

# (C) Processing -> Disposal (optional, but allows overflow if processing is capped)
for _, prow in proc_nodes.iterrows():
    arcs.append({
        "from": prow["node_id"],
        "to": DISPOSAL_NODE,
        "cost": DISPOSAL_COST,
        "cap": BIG_CAP
    })

# (D) Sorting -> Disposal (critical to guarantee feasibility)
for _, srow in sorting_nodes.iterrows():
    arcs.append({
        "from": srow["node_id"],
        "to": DISPOSAL_NODE,
        "cost": DISPOSAL_COST,
        "cap": BIG_CAP
    })

# ============================================================
# 3) Min-cost flow LP (now always feasible)
# ============================================================
n = len(arcs)
cvec = np.array([a["cost"] for a in arcs], dtype=float)
bounds = [(0.0, a["cap"]) for a in arcs]

nodes = list({a["from"] for a in arcs} | {a["to"] for a in arcs})
demand = {node: 0.0 for node in nodes}

# Supplies at collection nodes (negative demand in flow-balance form: inflow - outflow = demand)
for cnode in source_nodes:
    demand[cnode["node_id"]] = -float(region_supply[cnode["region"]])

# Disposal is sink that absorbs everything: set demand = TOTAL_SUPPLY
demand[DISPOSAL_NODE] = float(TOTAL_SUPPLY)

A_eq = []
b_eq = []
for node in nodes:
    row = np.zeros(n)
    for j, a in enumerate(arcs):
        if a["to"] == node:
            row[j] += 1.0
        if a["from"] == node:
            row[j] -= 1.0
    A_eq.append(row)
    b_eq.append(demand.get(node, 0.0))

A_eq = np.array(A_eq)
b_eq = np.array(b_eq)

res = linprog(cvec, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method="highs")
if not res.success:
    raise RuntimeError(f"Optimisation failed: {res.message}")

x_opt = res.x
opt_cost = float(res.fun)

# ============================================================
# 4) Baseline routing (simple equal split, with disposal fallback)
# ============================================================
baseline = np.zeros(n)

# Helper: even split with caps
def even_split(total, caps):
    caps = np.array(caps, dtype=float)
    if total <= 0 or len(caps) == 0:
        return np.array([])
    alloc = np.ones_like(caps) * (total / len(caps))
    # Cap and re-distribute iteratively
    for _ in range(20):
        over = alloc > caps
        if not over.any():
            break
        excess = (alloc[over] - caps[over]).sum()
        alloc[over] = caps[over]
        under = ~over
        if under.any():
            alloc[under] += excess / under.sum()
        else:
            break
    return np.minimum(alloc, caps)

# A) Region -> sorting
for cnode in source_nodes:
    r = cnode["region"]
    supply = float(region_supply[r])
    arcs_cs_idx = [i for i,a in enumerate(arcs) if a["from"]==cnode["node_id"]]
    caps = [arcs[i]["cap"] for i in arcs_cs_idx]
    alloc = even_split(supply, caps)
    baseline[arcs_cs_idx] = alloc

# B) Sorting -> processing + disposal
for _, srow in sorting_nodes.iterrows():
    s_id = srow["node_id"]
    inflow_s = baseline[[i for i,a in enumerate(arcs) if a["to"]==s_id]].sum()

    # outgoing processing arcs
    out_sp_idx = [i for i,a in enumerate(arcs) if a["from"]==s_id and a["to"].startswith("P")]
    out_disp_idx = [i for i,a in enumerate(arcs) if a["from"]==s_id and a["to"]==DISPOSAL_NODE]

    proc_caps = [arcs[i]["cap"] for i in out_sp_idx]
    proc_alloc = even_split(inflow_s, proc_caps)
    baseline[out_sp_idx] = proc_alloc

    # Any remainder goes to disposal (guaranteed feasible)
    remainder = inflow_s - proc_alloc.sum()
    if remainder > 1e-9 and len(out_disp_idx)==1:
        baseline[out_disp_idx[0]] = remainder

# C) Processing -> disposal (baseline pushes all processing output to disposal sink to satisfy flow balance)
# (Because disposal is the network sink in this formulation.)
for _, prow in proc_nodes.iterrows():
    p_id = prow["node_id"]
    inflow_p = baseline[[i for i,a in enumerate(arcs) if a["to"]==p_id]].sum()
    out_pdisp_idx = [i for i,a in enumerate(arcs) if a["from"]==p_id and a["to"]==DISPOSAL_NODE]
    if len(out_pdisp_idx)==1:
        baseline[out_pdisp_idx[0]] = inflow_p

base_cost = float(np.dot(baseline, cvec))

# ============================================================
# 5) OUTPUTS: baseline vs optimised + selective routing evidence
# ============================================================
# Build arc flow tables
flow_opt = pd.DataFrame(arcs)
flow_opt["flow"] = x_opt
flow_opt["type"] = np.where(flow_opt["to"]==DISPOSAL_NODE, "to_disposal", "internal")

flow_base = pd.DataFrame(arcs)
flow_base["flow"] = baseline
flow_base["type"] = np.where(flow_base["to"]==DISPOSAL_NODE, "to_disposal", "internal")

# Output 1: Cost comparison
cost_reduction_pct = (base_cost - opt_cost) / base_cost * 100 if base_cost > 0 else np.nan
print("Output 1: Baseline vs Optimised Total System Cost")
print(f"Baseline total cost : {base_cost:,.2f}")
print(f"Optimised total cost: {opt_cost:,.2f}")
print(f"Cost reduction (%)  : {cost_reduction_pct:.2f}%")

# Output 2: Disposal (landfill leakage proxy)
opt_disposal = flow_opt.loc[flow_opt["to"]==DISPOSAL_NODE, "flow"].sum()
base_disposal = flow_base.loc[flow_base["to"]==DISPOSAL_NODE, "flow"].sum()
print("\n Output 2: Disposal Flow (Landfill Leakage Proxy) ")
print(f"Baseline diverted to disposal : {base_disposal:,.2f} ({base_disposal/TOTAL_SUPPLY*100:.2f}%)")
print(f"Optimised diverted to disposal: {opt_disposal:,.2f} ({opt_disposal/TOTAL_SUPPLY*100:.2f}%)")

# Output 3: Facility utilisation (sorting + processing)
# Sorting utilisation: inflow to sorting vs sorting capacity
sort_cap = sorting_nodes.set_index("node_id")["current_capacity"].to_dict()
sort_in_opt = flow_opt[flow_opt["to"].isin(sort_cap.keys())].groupby("to")["flow"].sum()
sort_util_opt = pd.DataFrame({"capacity": pd.Series(sort_cap), "inflow": sort_in_opt}).fillna(0)
sort_util_opt["util_%"] = np.where(sort_util_opt["capacity"]>0, 100*sort_util_opt["inflow"]/sort_util_opt["capacity"], 0)
sort_util_opt = sort_util_opt.sort_values("util_%", ascending=False).reset_index().rename(columns={"index":"sorting_node"})

proc_cap = proc_nodes.set_index("node_id")["current_capacity"].to_dict()
proc_in_opt = flow_opt[flow_opt["to"].isin(proc_cap.keys())].groupby("to")["flow"].sum()
proc_util_opt = pd.DataFrame({"capacity": pd.Series(proc_cap), "inflow": proc_in_opt}).fillna(0)
proc_util_opt["util_%"] = np.where(proc_util_opt["capacity"]>0, 100*proc_util_opt["inflow"]/proc_util_opt["capacity"], 0)
proc_util_opt = proc_util_opt.sort_values("util_%", ascending=False).reset_index().rename(columns={"index":"processing_node"})

print("\n Output 3A: Sorting Facility Utilisation (Optimised) — Top 10 ")
print(sort_util_opt.head(10).to_string(index=False))

print("\n Output 3B: Processing Facility Utilisation (Optimised) — Top 10 ")
print(proc_util_opt.head(10).to_string(index=False))

# Output 4: Unused facilities (selective routing evidence)
unused_sort = (sort_util_opt["inflow"] <= 1e-9).sum()
unused_proc = (proc_util_opt["inflow"] <= 1e-9).sum()
print("\n Output 4: Selective Use of Infrastructure (Optimised) ")
print(f"Unused sorting facilities (in optimised solution): {unused_sort} out of {len(sort_util_opt)}")
print(f"Unused processing facilities (in optimised solution): {unused_proc} out of {len(proc_util_opt)}")

# Output 5: Top internal flows (shows targeted routing)
top_opt_flows = flow_opt[flow_opt["type"]=="internal"].sort_values("flow", ascending=False).head(15)
print("\n Output 5: Top 15 Internal Flow Allocations (Optimised) ")
print(top_opt_flows[["from","to","flow","cost"]].to_string(index=False))
```
